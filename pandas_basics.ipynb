# Calculation of Best App according to formula 
## Best_app = (rate*installs)/price

```python
price = data1['Price'].str.extract('(\d+)', expand=False).astype(float) + 1  

installs = data1['Installs'].str.replace('+','').str.replace(',','').str.replace('Free','Nan').astype(float)

Best_app = ( data1['Rating'] * installs) / price
Best_app

```

# Average Best_app score for each "Content Rating"

```python

import numpy as np

new_data = pd.concat([data1['Content Rating'],pd.DataFrame(Best_app,columns=['Score'])],axis=1)

average_rating = new_data.groupby(new_data['Content Rating'])['Score'].mean()
average_rating
```

# Bar Chart that plots scores against each content rating

```python
average_rating.plot(kind='bar')
```

# Merging two datasets

```python
merged_data = pd.merge(data1,data2,on='App')
merged_data
```

# Top 10 Applications according to No of reviews (Operations are performed on Merged Dataset)

```python

grouped_by_reviews_data = merged_data.groupby('App')['Reviews'].first().reset_index()

grouped_by_reviews_data['Reviews'] =  grouped_by_reviews_data['Reviews'].str.extract('(\d+)',expand=False).astype(float)

top_10_reviews = grouped_by_reviews_data.sort_values(['Reviews'], ascending = False)

top_10_reviews.head(10)
```

# Calculating Discussion Rate having formula
## Discussion_rate = (number_of_reviews*average_lenght_of_review)/installs

```python
merged_data['Length_Of_Reviews'] = merged_data['Translated_Review'].str.len()
merged_data['Price'] = merged_data['Price'].str.extract('(\d+)', expand=False).astype(float)

new_metrices = merged_data.groupby(['App']).aggregate({'Reviews' : 'first' , 'Length_Of_Reviews' : 'mean','Installs' : 'first', 'Price' : 'mean','Content Rating': 'first'}).reset_index()

new_metrices['Reviews'] =  new_metrices['Reviews'].str.extract('(\d+)',expand=False).astype(float)

new_metrices['Installs'] = new_metrices['Installs'].str.replace('+','').str.replace(',','').str.replace('Free','Nan').astype(float)

new_metrices['Discussion_Rate'] = ( new_metrices['Reviews'] * new_metrices['Length_Of_Reviews'] ) / new_metrices['Installs']

new_metrices
```

# Top 10 Applications according to Discussion Rate

```python
new_metrices = new_metrices.sort_values(["Discussion_Rate"], ascending = False).head(10)
new_metrices
```

# Bar chart that plots Price and Discussion Rate against Content Rating

```python
plot_metric = new_metrices.drop(['App', 'Reviews','Length_Of_Reviews','Installs'], axis=1)
plot_metric.head(10).plot( x = 'Content Rating', y = ['Price','Discussion_Rate'], kind = 'bar')
```
